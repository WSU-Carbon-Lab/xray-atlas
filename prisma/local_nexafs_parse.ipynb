```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
sp_data_files = (
    Path.home()
    / "Washington State University (email.wsu.edu)"
    / "Carbon Lab Research Group - Documents"
    / "NEXAFS Database Project"
)

molecules = [f.name for f in sp_data_files.iterdir() if f.name not in ["Chemical Structures", "Initial Materials"] and f.is_dir()]
_ = [print(f) for f in molecules]
```

```python
data_files = [f for f in sp_data_files.rglob("*.xlsx") if f.parent.name in molecules]
```

```python
import dis


for data in data_files:
    xlf = pd.ExcelFile(data)
    sheet_names = xlf.sheet_names

    sheets = []
    for sheet in sheet_names:
        df = xlf.parse(sheet)
        sheets.append(df)
    df = pd.concat(sheets, axis=1)
display(df)
df.describe()
```

```python
from math import remainder
import pyodbc

conn = pyodbc.connect(
    "DRIVER={ODBC Driver 18 for SQL Server};SERVER=HDUVALLH;DATABASE=NEXAFS;Trusted_Connection=yes;TrustServerCertificate=yes"
)

def get_exp_id(data, conn):
    # Locate unique identifiers in the columns
    unique = {id[:-3] for id in [col.split("deg")[0] for col in data.columns]}
    out = {unique: None for unique in unique}
    for id in unique:
        split = id.split("_")
        mol = split[0]
        atom = split[1]
        edge = split[2]
        nex_type = split[4]
        source = split[5]
        endstation = split[6]

        query = f"""
        SELECT ExperimentID
        FROM EXP_LIST
        WHERE name = '{mol}'
        AND atom = '{atom}'
        AND edge = '{edge}'
        AND NEXAFSMethod = '{nex_type}'
        AND source = '{source}'
        AND endstation = '{endstation}'
        """
        exp_id = conn.execute(query).fetchone()[0]
        out[id] = exp_id
    return out

key_pairs = get_exp_id(df, conn)
```

```python
deg = {col.split("deg")[0][-2:] for col in df.columns}

dfs = []
for d in deg:
    for key, id in key_pairs.items():
        e = df[f"{key}_{d}deg_En"].dropna()
        i = df[f"{key}_{d}deg"].dropna()
        deg = [d] * len(e)

        dfs.append(pd.DataFrame({"exp_id": key, "e": e, "mu": i, "deg": deg}))

new_df = pd.concat(dfs)
new_df
```

```python
def add_to_db(df, conn):
    for i, row in df.iterrows():
        id = row.exp_id
        split = id.split("_")
        mol = split[0]
        atom = split[1]
        edge = split[2]
        nex_type = split[4]
        source = split[5]
        endstation = split[6]

        query = f"""
        DECLARE @exp_id UNIQUEIDENTIFIER

        SELECT @exp_id = ExperimentID
        FROM EXP_LIST
        WHERE name = '{mol}'
        AND atom = '{atom}'
        AND edge = '{edge}'
        AND NEXAFSMethod = '{nex_type}'
        AND source = '{source}'
        AND endstation = '{endstation}'

        INSERT INTO Nexafs (exp_id, e, mu, deg)
        VALUES (@exp_id, {row.e}, {row.mu}, {row.deg})
        """
        conn.execute(query)
    conn.commit()

add_to_db(new_df, conn)
```
